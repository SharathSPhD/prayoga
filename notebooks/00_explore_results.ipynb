This notebook pulls **prayoga's committed aggregate results** straight from GitHub and reproduces every headline figure. No model, no GPU, no installs — just `Runtime > Run all`.

**Thesis:** LLM jailbreak, hypnotic suggestion, and tantric *vaśīkaraṇa* are three instances of one mechanism — suppress the monitoring/refusal faculty, co-opt automatic generation. Honest hardened headline: *asymmetry with a shared necessary core*.

```python
import json, urllib.request
import matplotlib.pyplot as plt, numpy as np
RAW = 'https://raw.githubusercontent.com/SharathSPhD/prayoga/main'
def load(p):
    return json.loads(urllib.request.urlopen(f'{RAW}/{p}').read())
findings = load('site/public/data/findings.json')['findings']
M = load('site/public/data/metrics.json')
print(f'{len(findings)} findings; metrics: {list(M.keys())}')
```

## F2 — Refusal abliteration dose–response (EC50)

```python
d = M['dose']; a = np.array(d['alphas'])
plt.figure(figsize=(6,4))
plt.scatter(a, d['asr_real'], color='#c0392b', zorder=3, label='refusal direction')
plt.scatter(a, d['asr_random'], color='#7f8c8d', marker='s', label='random direction')
plt.axvline(d['ec50'], ls='--', color='#2c3e50'); plt.text(d['ec50']+.02,.1,f"EC50={d['ec50']:.2f}")
plt.xlabel('ablation strength α'); plt.ylabel('attack success rate'); plt.legend()
plt.title(f"Dose–response (R²={d['r2']:.3f})"); plt.show()
```

## F18 — Effective dimension is **layer-dependent** (qualifies F8)

```python
plt.figure(figsize=(6,4))
for mdl, c in [('gemma-2-2b-it','#c0392b'),('qwen2.5-3b-it','#2980b9'),('gemma-2-9b-it','#27ae60')]:
    if mdl in M.get('dimsweep',{}):
        s = M['dimsweep'][mdl]; plt.plot(s['layers'], s['dim'], 'o-', ms=3, color=c, label=mdl)
plt.xlabel('layer'); plt.ylabel('refusal-subspace effective dim')
plt.title('Dimension varies across layers — the clean per-model contrast is layer-specific')
plt.legend(); plt.show()
```

## F11 — Refusal is a paraphrase-orbit invariant; injection breaks it

```python
sym = M['symmetry']; models = list(sym.keys()); x = np.arange(len(models)); w=.35
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,3.6))
ax1.bar(x-w/2,[sym[m]['F_refusal'] for m in models],w,color='#c0392b',label='refusal dir')
ax1.bar(x+w/2,[sym[m]['F_random'] for m in models],w,color='#7f8c8d',label='random')
ax1.set_yscale('log'); ax1.set_xticks(x); ax1.set_xticklabels(models,fontsize=8); ax1.set_ylabel('orbit F-ratio (log)'); ax1.legend(); ax1.set_title('orbit invariance')
ax2.bar(x-w/2,[sym[m]['m_plain'] for m in models],w,color='#2980b9',label='plain')
ax2.bar(x+w/2,[sym[m]['m_injected'] for m in models],w,color='#e67e22',label='injected')
ax2.set_xticks(x); ax2.set_xticklabels(models,fontsize=8); ax2.set_ylabel('order parameter m'); ax2.legend(); ax2.set_title('injection breaks it'); plt.show()
```

## ṣaṭkarma — the six-act intervention taxonomy

```python
sk = M['satkarma']
names=[r['sanskrit'] for r in sk]; eff=[r['effect'] for r in sk]; ctl=[r['control'] for r in sk]
y=np.arange(len(names))[::-1]; h=.36
plt.figure(figsize=(7,4))
plt.barh(y+h/2,eff,h,color=['#c0392b' if r['separated'] else '#bbb' for r in sk],label='effect')
plt.barh(y-h/2,ctl,h,color='#7f8c8d',label='control')
plt.yticks(y,names); plt.xlabel('measured effect'); plt.legend(); plt.title('ṣaṭkarma as activation interventions'); plt.show()
```
