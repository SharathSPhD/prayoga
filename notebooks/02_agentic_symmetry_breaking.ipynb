## Redacted multi-turn trajectory

A live white-box run would capture residual activations at each turn and project them onto the refusal direction. Public demos should publish only redacted turn labels and aggregate metrics.

```python
turns = np.array([0, 1, 2])
labels = ['setup', 'escalate', 'target']
order_parameter = np.array([0.153, 0.128, 0.101])
precision_proxy = np.array([7.18, 4.72, 3.14])

fig, ax1 = plt.subplots(figsize=(7, 4))
ax1.plot(turns, order_parameter, 'o-', label='order parameter m')
ax1.set_xticks(turns, labels)
ax1.set_ylabel('refusal order parameter')
ax2 = ax1.twinx()
ax2.plot(turns, precision_proxy, 's--', color='tab:orange', label='precision proxy')
ax2.set_ylabel('monitoring-precision proxy')
ax1.set_title('Redacted trajectory: refusal projection drops across turns')
fig.tight_layout()
```
